# Get the World Bank Data

```python
from pandas_datareader import wb
```

```python
dat = wb.download(indicator='NY.GDP.PCAP.KD', country=['all'], start=2003, end=2016)
```

```python
dat.info()
```

```python
dat.head()
```

# Reset the multi-index

```python
df_gdp_years = dat.reset_index()
```

```python
df_gdp_years.head(10)
```

# Pivot Table

```python
df_gdp_ready = df_gdp_years.pivot_table(index= 'country', columns='year', values="NY.GDP.PCAP.KD")
```

```python
df_gdp_ready.head()
```

```python
df_gdp_ready.to_csv('df_gdp_ready.csv', sep='\t')
```
